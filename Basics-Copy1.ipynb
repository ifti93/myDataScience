### Instruction: Any examples of numeric data stored as text needs to be cleaned (it must be in numeric form)

```python
#display pricing (defined as text/non-numeric initiated with "$"). 
#Our objective is to convert them to numeric form like int or float datatypes.
autos["price"]
```

```python
autos["price"].values
```

```python
autos["price"]= autos["price"].str.replace("$",'')
```

```python
autos["price"].values
```

```python
autos["price"]= autos["price"].astype(int)  
#Error: wrong way to convert price to int (numeric) from string/text (non numeric)
```

```python
# Correct way to do conversion/parsing datatype from text/string to int 
#replace([list of things to match],[if yes,do this],[enable "regex" by typing "True])
#modified price values by removing $ & , like $5,000 & then converting/parsing it to int or float (float is like int but with decimal support)
autos["price"]= autos["price"].replace('[$,]','',regex=True).astype(int) 
```

```python
# Display the prices in numeric form.
autos["price"]
```

```python
autos["odometer"] 
#display values in one line you can also display dataframe "autos" like autos[column name].values
#As you see problem is data is not cleaned as it's showing km so its in string/text/non-numeric format, our goal is to convert
# it to numeric form , int would work so as float although if digits are not important, then better option is int due to 
# less memory/space allocation.
```

```python
# Successful conversion/parsing from string to int
autos["odometer"]= autos["odometer"].replace("[,km]",'',regex=True) 
autos["odometer"]
```

```python
autos["odometer"].astype(int)
```
